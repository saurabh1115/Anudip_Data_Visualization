1. Analyze the relationship between the size of houses (measured in square footage) and their selling prices in a particular neighborhood. You have collecteddata on various houses in that neighborhood.Create a scatter plot using thebelow data and shape your conclusion/analysis.

```python
import numpy as np
import matplotlib.pyplot as plt
square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800,3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])
plt.figure(figsize=(10, 6))
plt.scatter(square_footage, selling_prices, color='green', marker='o')
plt.title('Relationship between House Size and Selling Price')
plt.xlabel('Square Footage(sq.ft)')
plt.ylabel('Selling Price ($)')
plt.grid(True)
plt.show()
```

2. Create a pie chart to visualize the distribution of your monthly income by source. You have collected data on the various sources of your income, such as salary, freelance work, investments, and rental income. Share yourconclusion/analysis.

```python
income_sources = ['Salary', 'Freelance', 'Investments', 'Rental', 'Other']
monthly_income = [5000, 1500, 1000, 600, 400]
plt.figure(figsize=(8, 8))
plt.pie(monthly_income, labels=income_sources, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Distribution of Monthly Income by Source')
plt.show()
```

3. Create a pie chart to illustrate the distribution of a company's revenue across its various business segments. You have collected data on the revenue generated by each segment, such as Product A, Product B, Services, and Licensing. Share your conclusion/analysis.

```python
segments = ['Product A', 'Product B', 'Services', 'Licensing']
revenue_percentages = [45, 25, 15, 15]
plt.figure(figsize=(8, 8))
plt.pie(revenue_percentages, labels=segments, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Revenue Distribution by Business Segment')
plt.show()
```

4. Suppose you're a sales manager for an e-commerce company, and you want to create a figure with subplots to compare the sales performance of different product categories over time. You have sales data for four product categories: Electronics, Clothing, Home & Garden, and Sports & Outdoors. Share your conclusion/analysis.

```python
months = np.arange(1, 13)
electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000,
36000, 38000, 39000, 41000, 42000])
clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000,
23000, 24000, 25000, 26000])
home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000,
25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000,
19000, 20000, 21000, 22000, 23000])
# Create figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Plot for Electronics
axs[0, 0].plot(months, electronics_sales, marker='o', color='b')
axs[0, 0].set_title('Electronics Sales')
axs[0, 0].set_xlabel('Months')
axs[0, 0].set_ylabel('Sales Amount')
axs[0, 0].grid(True)

# Plot for Clothing
axs[0, 1].plot(months, clothing_sales, marker='o', color='g')
axs[0, 1].set_title('Clothing Sales')
axs[0, 1].set_xlabel('Months')
axs[0, 1].set_ylabel('Sales Amount')
axs[0, 1].grid(True)

# Plot for Home & Garden
axs[1, 0].plot(months, home_garden_sales, marker='o', color='r')
axs[1, 0].set_title('Home & Garden Sales')
axs[1, 0].set_xlabel('Months')
axs[1, 0].set_ylabel('Sales Amount')
axs[1, 0].grid(True)

# Plot for Sports & Outdoors
axs[1, 1].plot(months, sports_outdoors_sales, marker='o', color='y')
axs[1, 1].set_title('Sports & Outdoors Sales')
axs[1, 1].set_xlabel('Months')
axs[1, 1].set_ylabel('Sales Amount')
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

1. Visualize the daily temperature changes over time in a city and give your conclusion

```python
import matplotlib.pyplot as plt
# Sample data: Daily temperatures for a month
days = list(range(1, 32))
temperatures = [65, 68, 70, 72, 75, 76, 78, 80, 81, 79, 75, 72, 70, 68, 67, 69, 70, 73, 75, 76, 78, 80, 81, 82, 83, 82, 80, 78, 76, 74, 71]
# Create the line plot
plt.plot(days, temperatures,marker='o')
# Add labels and a title
plt.xlabel("Day of the Month")
plt.ylabel("Temperature (°F)")
plt.title("Daily Temperature Over a Month")
plt.show()
```

2. Create a line plot to visualize the daily closing prices of a stock over a year and give your conclusion.

```python
import matplotlib.pyplot as plt

# Data
days = list(range(1, 78))
stock_prices = [100, 105, 110, 115, 112, 120, 118, 125, 128, 130, 132, 135, 138, 140, 142, 144, 145, 148, 150, 155, 160, 158, 162, 165, 170,
                172, 175, 178, 180, 182, 185, 188, 190, 192, 195, 198, 200, 198, 195, 193, 190, 188, 185, 182, 180, 178, 175, 172, 170, 168, 
                165, 162, 160, 158, 155, 152, 150, 148, 145, 143, 140, 138, 135, 132, 130, 128, 125, 123, 120, 118, 115, 112, 110, 108, 105, 
                103, 100]

# Plotting the data
plt.figure(figsize=(12, 6))
plt.plot(days, stock_prices, marker='o', linestyle='-', color='blue')
plt.title('Daily Closing Prices of a Stock Over 78 Days')
plt.xlabel('Days')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.xticks([0, 10, 20, 30, 40, 50, 60, 70, 80])
plt.show()
```
